```python
#Qustion No:1
import numpy as np
import pandas as pd
df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN', 'londON_StockhOlm', 
 
                              'Budapest_PaRis', 'Brussels_londOn'], 
 
             'FlightNumber': [10045, np.nan, 10065, np.nan, 10085], 
 
             'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]], 
 
                  'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )', 
 
                             '12. Air France', '"Swiss Air"']})

print("\nThe Original DataFrame\n","-"*75,sep="")
print(df)
df=df.fillna(0)
print("\nFilling Nan with 0\n","-"*75,sep="")
print(df)
print("\nThe FlightNumber column with integer values\n","-"*75,sep="")
L=list(range(10045,10095,10))
df['FlightNumber']= L
df
```
